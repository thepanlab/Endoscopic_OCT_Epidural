# Resnet 50

```python
import pandas as pd
import numpy as np
import pickle
```

```python
import os, sys
# Open a file
path = "/scratch/paulcalle/Epidural_results/archRESNET50_results/"
dirs = os.listdir( path )
```

```python
dirs
```

```python
import re

l_history_files = []

r = re.compile('history')

for item in dirs:

    res = r.search(item)
    if res != None:
        l_history_files.append(item)
```

```python
l_history_files
```

```python
df_val_results_resnet50 = pd.DataFrame(np.nan, index=["K%s_outer"%i for i in np.arange(1,9,1)],
                     columns=["k%s_val"%i for i in np.arange(1,9,1)])

max_epoch = 0 

for item in l_history_files:
    print(item)
    # Get outter look
    r_outer = re.compile(r'(?<=K)[0-9]+')
    k_outer = r_outer.search(item).group(0)
    print("k_outer = ", k_outer)

    # Get inner look
    r_inner = re.compile(r'(?<=k)[0-9]+')
    k_val = r_inner.search(item).group(0)
    print("k_val = ", k_val)
    
    with open('/scratch/paulcalle/Epidural_results/archRESNET50_results/' + item, 'rb') as handle:
        history_temp = pickle.load(handle)
    
    # Search index of validation
    ind_min_val_loss = np.argmin(history_temp['val_loss'])
    if ind_min_val_loss > max_epoch:
        max_epoch = ind_min_val_loss
    
    print("ind_min_val_loss = ", ind_min_val_loss)
    val_acc = history_temp['val_accuracy'][ind_min_val_loss]
    print("val_acc = ", val_acc)
    
    df_val_results_resnet50.loc["K%s_outer"%k_outer,"k%s_val"%k_val] = val_acc
    
print("max epoch = ",max_epoch)
```

# MobileNetv2

```python
import pandas as pd
import numpy as np
import pickle
```

```python
import os, sys
# Open a file
path = "/scratch/paulcalle/kidney_results/arch_mobilenetv2_results"
dirs = os.listdir( path )
```

```python
dirs
```

```python
import re

l_history_files = []

r = re.compile('^history')

for item in dirs:

    res = r.search(item)
    if res != None:
        l_history_files.append(item)
```

```python
l_history_files
```

```python
import matplotlib.pyplot as plt
```

```python
df_val_results_mobilenetv2 = pd.DataFrame(np.nan, index=["K%s_outer"%i for i in np.arange(1,11,1)],
                     columns=["k%s_val"%i for i in np.arange(1,11,1)])



for item in l_history_files:
    print(item)
    # Get outter look
    r_outer = re.compile(r'(?<=K)[0-9]+')
    k_outer = r_outer.search(item).group(0)
    print("k_outer = ", k_outer)

    # Get inner look
    r_inner = re.compile(r'(?<=k)[0-9]+')
    k_val = r_inner.search(item).group(0)
    print("k_val = ", k_val)
    
    with open('/scratch/paulcalle/kidney_results/arch_mobilenetv2_results/' + item, 'rb') as handle:
        history_temp = pickle.load(handle)
    
    # Search index of validation

    ind_min_val_loss = np.argmin(history_temp['val_loss'])
    
#     if ind_min_val_loss <= 10:
#         plt.plot( history_temp['val_loss'])

    plt.plot( history_temp['val_loss'])


    print("ind_min_val_loss = ", ind_min_val_loss)
    val_acc = history_temp['val_accuracy'][ind_min_val_loss]
    print("val_acc = ", val_acc)
    
    df_val_results_mobilenetv2.loc["K%s_outer"%k_outer,"k%s_val"%k_val] = val_acc
    
plt.show()
```

```python
df_val_results_mobilenetv2_tf = pd.DataFrame(np.nan, index=["K%s_outer"%i for i in np.arange(1,11,1)],
                     columns=["k%s_val"%i for i in np.arange(1,11,1)])

for item in l_history_files:
    print(item)
    # Get outter look
    r_outer = re.compile(r'(?<=K)[0-9]+')
    k_outer = r_outer.search(item).group(0)
    print("k_outer = ", k_outer)

    # Get inner look
    r_inner = re.compile(r'(?<=k)[0-9]+')
    k_val = r_inner.search(item).group(0)
    print("k_val = ", k_val)
    
    with open('/scratch/paulcalle/kidney_results/arch_mobilenetv2_tf_results/' + item, 'rb') as handle:
        history_temp = pickle.load(handle)
    
    # Search index of validation
    ind_min_val_loss = np.argmin(history_temp['val_loss'])
    
    plt.plot( history_temp['val_loss'])
    
    print(history_temp['val_loss'])
    print("ind_min_val_loss = ", ind_min_val_loss)
    val_acc = history_temp['val_accuracy'][ind_min_val_loss]
    print("val_acc = ", val_acc)
    
    df_val_results_mobilenetv2_tf.loc["K%s_outer"%k_outer,"k%s_val"%k_val] = val_acc
```
